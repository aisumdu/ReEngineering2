Finally, we can get started with the actual git stuff. First thing we need to do is initialize a local git repository (i.e in our google drive project folder)

```
!git init
```

Now create a private git repository. I named mine as “gitTutorial” and then we need to add this repository as a remote repository to our git project.

!git remote add origin https://{github_username}:{github_password}@github.com/{github_username}/{github_repo_name}.git

Replace the placeholder with your detail. For example if my github username was saifullah73, my password was 12345 and my repository name is gitTutorial so the above command,for me, would be “!git remote add origin https://saifullah73:12345@github.com/saifullah73/GitTutorial.git”

```
!git remote add origin https://aisumdu:aisumdu2020@github.com/aisumdu/ReEngineering2.git
```
